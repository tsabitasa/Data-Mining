```python
# Fungsi untuk mengubah tipe data
def change_data_type(df, column_types):
    for column, dtype in column_types.items():
        try:
            if dtype == 'int' or dtype == 'Int64':
                # Ekstrak angka dari string dan ubah ke numerik
                df[column] = df[column].astype(str).str.extract('(\d+)')[0]  # Ekstrak hanya angka
                df[column] = pd.to_numeric(df[column], errors='coerce').astype('Int64')  # ubah ke Int64
                print(f"Tipe data kolom '{column}' berhasil diubah menjadi {dtype}.")
            else:
                # Lakukan konversi standar jika bukan integer
                df[column] = df[column].astype(dtype)
                print(f"Tipe data kolom '{column}' berhasil diubah menjadi {dtype}.")
        except ValueError as e:
            print(f"Error mengubah kolom '{column}' ke tipe {dtype}: {e}")

    return
```

```python
# Fungsi untuk menghitung nilai unik pada setiap variabel
from collections import Counter
def count_all_values(df):
    result = {}

    for column in df.columns:
        value_counts = Counter(df[column])
        result[column] = value_counts

    return result
```

```python
# Fungsi untuk menghapus kelebihan spasi pada entri
def remove_extra_spaces(df):
    # Simpan tipe data asli setiap kolom sebelum diubah
    original_types = df.dtypes.to_dict()

    # Hapus spasi berlebih di setiap kolom yang bertipe object atau category
    for column in df.columns:
        if df[column].dtype in ['object', 'category']:  # Cek apakah tipe data object atau category
            # Ubah sementara kolom category ke object
            if df[column].dtype == 'category':
                df[column] = df[column].astype('object')

            # Pastikan semua nilai adalah string sebelum menggunakan .str
            df[column] = df[column].astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)

    # Kembalikan tipe data kolom ke tipe aslinya
    for column, dtype in original_types.items():
        try:
            # Jika tipe asli adalah category, kita buat kategori ulang dengan nilai unik yang sudah dibersihkan
            if dtype.name == 'category':
                df[column] = pd.Categorical(df[column])
            else:
                df[column] = df[column].astype(dtype)
        except TypeError:
            print(f"Peringatan: Tidak dapat mengembalikan kolom '{column}' ke tipe {dtype}")

    return
```

```python
# Menghapus kelebihan spasi pada entri
remove_extra_spaces(df)
```

```python
# Fungsi untuk mengkategorikan variabel 'Waktu Mendapatkan Pekerjaan'
def categorize_job_search(value):
    # Daftar nilai untuk masing-masing kategori
    less_than_3_months = {'Kurang dari 3 bulan', '1', '2'}
    three_to_six_months = {'3', '4', '5', '6', '3-6 Bulan'}
    six_to_twelve_months = {'7', '8', '9', '10', '11', '12', '6-12 bulan'}
    more_than_twelve_months = {'>12 bulan', 'lebih dari 12 bulan', '12'}

    # Kategori yang tidak berubah
    if value in ['Belum mendapatkan pekerjaan', 'Tidak berencana mencari pekerjaan']:
        return value

    # Cek kategori berdasarkan set
    if value in less_than_3_months:
        return '<3 bulan'
    elif value in three_to_six_months:
        return '3-6 bulan'
    elif value in six_to_twelve_months:
        return '6-12 bulan'
    elif value in more_than_twelve_months:
        return '>12 bulan'

# Menerapkan fungsi ke kolom Waktu Mendapatkan Pekerjaan
df['Waktu Mendapatkan Pekerjaan'] = df['Waktu Mendapatkan Pekerjaan'].apply(categorize_job_search)

# Mengubah tipe data variabel 'Waktu Mendapatkan Pekerjaan' menjadi category
change_data_type(df, {'Waktu Mendapatkan Pekerjaan': 'category'})

# Melihat nilai unik
Counter(df['Waktu Mendapatkan Pekerjaan'])
```

```python
# Fungsi untuk mendeteksi outlier
def detect_outliers(df, column):

    Q1 = df[column].quantile(0.25)
    Q2 = df[column].quantile(0.50)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    print("Q1={}, Q3={}, IQR={}".format(Q1, Q3, IQR))

    bawah = (df[column] < (Q1 - 1.5 *IQR)) # Outlier bawah
    atas = (df[column]  > (Q3 + 1.5 *IQR)) # Outlier atas
    outliers_bawah = df.loc[bawah]
    outliers_atas = df.loc[atas]

    no_outlier = (df[column] >= Q1 - 1.5 * IQR) & (df[column] <= Q3 + 1.5 *IQR)
    df_no_outlier = df[no_outlier]
    df_no_outlier.reset_index(drop=True, inplace=True)

    return df_no_outlier, outliers_bawah, outliers_atas
```

```python
# Mendeteksi outlier pada variabel 'Penghasilan'
df_no_outlier, penghasilan_rendah, penghasilan_tinggi = detect_outliers(df, 'Penghasilan')
```

```python
# Menambahkan kolom 'Kategori Penghasilan' ke DataFrame
df['Kategori Penghasilan'] = 'Rata-rata'

# Menandai data dengan penghasilan tinggi
df.loc[df.index.isin(penghasilan_tinggi.index), 'Kategori Penghasilan'] = 'Tinggi'

# Menandai data dengan penghasilan rendah
df.loc[df.index.isin(penghasilan_rendah.index), 'Kategori Penghasilan'] = 'Rendah'

# Melihat distribusi kategori penghasilan
print(df['Kategori Penghasilan'].value_counts())
```

## **Rekomendasi**



**1.   Meningkatkan Dukungan Akademik**

Fakultas Sains dan Teknologi bisa memberikan bimbingan atau program remedial bagi alumni yang menyelesaikan studi lebih dari 6 tahun agar mereka dapat lulus tepat waktu. Selain itu, bisa juga dengan menyediakan program mentoring untuk membantu mereka menyelesaikan studi dengan baik. Program bimbingan dan mentoring dapat membantu mahasiswa yang tertinggal dan mencegah dropout, namun membutuhkan tenaga pengajar tambahan dan komitmen tinggi dari mahasiswa. Serta beban sumber daya juga bisa menjadi masalahnya, terutama jika mahasiswa tidak memanfaatkan bimbingan ini secara optimal.


**2. Program Pelatihan Karier**

Fakultas seharusnya sering mengadakan pelatihan karier yang lebih lengkap, mencakup keterampilan soft skills, workshop kewirausahaan, dan kesempatan networking untuk meningkatkan peluang kerja bagi alumni baru. Pelatihan karier dapat meningkatkan keterampilan dan peluang kerja mahasiswa serta membuka peluang networking. Namun, program ini membutuhkan biaya dan sumber daya tambahan, serta bisa kurang efektif jika mahasiswa tidak antusias atau tidak menganggapnya begitu penting.

**3. Kemitraan dengan Industri dan Peningkatan Kolaborasi**

Fakultas dapat membangun kemitraan stategis dengan perusahaan di bidang terkait untuk menyediakan tempat magang dan peluang kerja bagi mahasiswa, memberikan mereka pengalaman nyata sebelum memasuki dunia kerja. Untuk mendukung mahasiswa yang berminat beriwirausaha dapat disediakan program pelatihan khusus yang meliputi dasar-dasar bisnis, management keuangan, dan pengembangan produk. Selain itu, fakultas juga bisa menyediakan inkubator bisnis atau program kewiraushaan di kampus. Kolaborasi dengan industri dapat memberikan pengalaman yang relevan dan peluang untuk magang, namun proses membangun hubungan dengan industri memakan waktu dan biaya yang cukup banyak. Tidak semua mahasiswa mungkin tertarik dengan bidang kewirausahaan atau pekerjaan formal di bidang-bidang tertentu.

**4. Peningkatan Kesesuaian Kurikulum dalam Bidang Ilmu Kerja**

Fakultas melakukan evaluasi rutin terhadap kurikulum dan sering melibatkan alumni dalam proses pembaruan kurikulum untuk memastikan dan memperbarui isi materi ajar sesuai kebutuhan industri. Selain itu, fakultas bisa sering melakukan survei berkala terhadap alumni yang sedang mencari pekerjaan untuk memahami kebutuhan dan tantangan yang dihadapinya, sehingga fakultas dapat lebih menyesuaikan program pembelajaran dan dukungan yang diberikan. Kurikulum yang relevan dengan industri membuat lulusan lebih siap kerja, tetapi pembaruan kurikulum membutuhkan proses yang cukup lama dan tidak semua mahasiswa ingin langsung bekerja karena beberapa lebih tertarik untuk melanjutkan studi.

**5. Memperkuat Program Studi Lanjut**

Fakultas dapat memfasilitasi program-program untuk persiapan studi lanjut, seperti memberikan informasi dan akses ke program beasiswa serta kemudahan dalam pendaftaran di universitas ternama. Juga memberikan bimbingan mahasiswa dan pelatihan keterampilan akademik seperti penulisan dan penelitian. Dukungan bagi mahasiswa yang ingin melanjutkan pendidikan memperbesar peluang diterima di universitas ternama, namun memerlukan tenaga pengajar kompeten dan bisa kurang diminati oleh mereka yang lebih ingin langsung bekerja.

**6. Meningkatkan Jaringan Alumni**

Fakultas juga dapat mewadahi untuk membangun jaringan alumni yang kuat untuk memberikan dukungan dan berbagi informasi untuk peluang pekerjaan, serta menciptakan forum diskusi bagi alumni-alumni untuk saling berbagi pengalaman dan sumber daya kepada seluruh mahasiswa. Jaringan alumni yang kuat mendukung solidaritas dan peluang pekerjaan, tetapi memerlukan komitmen dan sumber daya untuk dikelola dengan baik. Alumni yang sibuk mungkin kurang aktif, dan menjaga hubungan tetap lancar dalam jangka panjang menjadi tantangan masalahnya.
